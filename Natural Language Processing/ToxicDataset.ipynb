```python
from nltk.tokenize import word_tokenize
import pandas as pd
import math
```

```python
df = pd.read_csv('../data/Toxic/train.csv')
df.head()
```

- limpar texto
- pegar palavras mais repetidas

```python
def create_lexicon(dataframe):
    lexicon = " "
    for i in range(math.floor(len(dataframe)/50)):
        lexicon += dataframe.get_value(i, 'comment_text')
    lexicon = word_tokenize(lexicon)
    return lexicon
```

```python
lexicon = create_lexicon(df)
lexicon
```

```python
import seaborn as sns
from collections import Counter
```

```python
def repeated_words(lex, n_of_occurences):
    w_counts = Counter(lex)
    w_counts_list = []
    w_list = []
    for w in w_counts:
        if w_counts[w]>n_of_occurences:
            w_list.append(w)
            w_counts_list.append(w_counts[w])
    return w_list, w_counts_list
```

```python
a, b = repeated_words(lexicon, 1000)
```
