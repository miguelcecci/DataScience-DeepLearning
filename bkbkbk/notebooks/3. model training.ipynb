importando o modulo de fun'coes do bk

```python
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
import imp

bk = imp.load_source('bk.py', '../bk.py')
```

```python
data = bk.get_training_dataset()
```

```python
data['Labels'].value_counts()
```

fazer downsampling para equilibrar os dados

### Balancing Labels

```python
length_of_data = len(data[data['Labels'] == 1])
```

```python
data = pd.concat([data[data['Labels'] == 1], data[data['Labels'] == 0].head(length_of_data)])
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data['Labels'].value_counts()
```

```python
labels = data['Labels'].values
del data['Labels']
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.33, random_state=42)
```

### CatBoostClassifier

```python
model = CatBoostClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
bk.compute_metrics(y_test, predictions)
```
