```python
import cv2 as cv
import tensorflow as tf
import matplotlib.pyplot as plt
import math
import random
%matplotlib inline
```

```python
test = cv.imread("../Screen Shot 2018-03-16 at 15.48.08.png", 0)
```

```python
fig = plt.figure(figsize=(18, 16), dpi = 80, facecolor='w', edgecolor='k')
plt.imshow(test)
```

```python
w, h = test.shape
```

```python
rw, rh = math.floor(w/3), math.floor(h/3)
```

```python
test_resized = cv.resize(test, (rw, rh))
fig = plt.figure(figsize=(18, 16), dpi = 80, facecolor='w', edgecolor='k')
plt.imshow(test_resized)
```

```python
test_blured = cv.resize(test_resized, (w, h))
fig = plt.figure(figsize=(18, 16), dpi = 80, facecolor='w', edgecolor='k')
plt.imshow(test_resized)
```

```python
wsub, hsub, test_pieces, blur_pieces = 5, 12, [], []
wsubsize, hsubsize = math.floor(w/wsub), math.floor(h/hsub)
for i in range(hsub-1):
    for j in range(wsub-1):
        test_pieces.append(test[hsubsize*i:hsubsize*(i+1), wsubsize*j:wsubsize*(j+1)])
        blur_pieces.append(test_blured[hsubsize*i:hsubsize*(i+1), wsubsize*j:wsubsize*(j+1)])
        
```
